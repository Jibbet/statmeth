The expectation value of the uniform distribution is
\begin{align}
    E\left[X\right]&=\int_a^b \frac{x}{b-a}\;dx
    \\&=\frac{1}{2}\frac{b^2-a^2}{b-a}
    \\&=\frac{a+b}{2}
\end{align}
For $b=1$ and $a=0$ this is $E\left[X\right]=0.5$.
The Variance of the uniform distribution is
\begin{align}
    V\left[X\right]&=\int_a^b \frac{\left(X-E\left[X\right]\right)^2}{b-a}\;dx
    \\&=\int_a^b \frac{\left(X-\frac{a+b}{2}\right)^2}{b-a}\;dx
    \\&=\frac{1}{b-a}\frac{1}{3}\left[\left(x-\frac{a+b}{2}\right)^3\right]_a^b
    \\&=\frac{1}{b-a}\frac{1}{3}\left[\left(b-\frac{a+b}{2}\right)^3-\left(a-\frac{a+b}{2}\right)^3\right]
    \\&=\frac{1}{b-a}\frac{1}{3}\frac{1}{8}\left[\left(b-a\right)^3-\left(a-b\right)^3\right]
    \\&=\frac{\left(b-a\right)^3}{12}
\end{align}
For $b=1$ and $a=0$ this is $V\left[X\right]=0.083$ $\rightarrow \sigma \approx 0.289$


```python
import numpy as np

a=0
b=1
N=np.arange(100,100000,100)

means, deviations = [],[]

for n in N:
    sample = np.random.uniform(a,b,n)
    means += [np.mean(sample)]
    deviations+= [np.std(sample)]
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(2,1,constrained_layout=True)
ax[0].axhline(y = 0.5, color = "r")
ax[0].scatter(N,means, s=1)
ax[0].set_xlabel("N")
ax[0].set_ylabel("sample mean")

v = np.sqrt(1/12)
ax[1].axhline(y = v, color = "r")
ax[1].scatter(N,deviations, s=1)
ax[1].set_xlabel("N")
ax[1].set_ylabel("sample standard deviation")
```
