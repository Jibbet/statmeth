```python
# import libraries
import pandas as pd  # useful for processing csv files
import matplotlib.pyplot as plt  # useful for plotting
```

```python
# import csv file with pandas
data = pd.read_csv('ClimateData.csv', sep=';', index_col='Decade')
```

```python
# show head of csv file
data.head()
```

```python
# plot the time series

##############
#YOUR CODE HERE
##############
dec = data.index.values
temp = data.GlobalAvgTemp.values
pir = data.NmbrPirates.values
co = data.Co2Concentration.values
fix, ax = plt.subplots(3,1,constrained_layout=True)
ax[0].scatter(dec, temp, s=1)
ax[0].set_ylabel("global average temperature")
ax[0].set_xlabel("decade")

ax[1].scatter(dec, pir, s=1)
ax[1].set_ylabel("number of pirates")
ax[1].set_xlabel("decade")

ax[2].scatter(dec, co, s=1)
ax[2].set_ylabel("Co2 concentration")
ax[2].set_xlabel("decade")
##############
```

```python
# calculate correlation coefficient
# you may use any library functions for this! 
# (you don't have to write the calculation yourself)

##############
#YOUR CODE HERE
##############
cr = data.corr()
cr.head()
##############
```
